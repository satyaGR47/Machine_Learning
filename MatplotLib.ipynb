## MatplotLib

Matplotlib is a plotting library for the Python programming language and it is numerical mathematics extension of Numpy.
It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of MatplotLib are:

- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-Quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline #-->we don't need to write plt.show() if we will use this modules
```

```python
import numpy as np
```

```python
## simple example

x= np.arange(0,10)
y= np.arange(11,21)
```

```python
plt.scatter(x,y,c = 'b')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title(" Scatter graph of x and y value")
```

```python
plt.plot(x,x*x,"r--*") #it will show the straight line
```

```python
plt.subplot(2,2,1) #-->it will draw the subdiagram
plt.plot(x,x*x,"r")
plt.subplot(2,2,2)
plt.plot(x,x*x,"b")
plt.subplot(2,2,3)
plt.plot(x,x*x,"g")
plt.subplot(2,2,4)
plt.plot(x,x*x,'y')
```

```python
#Bar plot

x = [2,8,10]
y = [11,22,16]

x2 = [5,9,10]
y2 = [10,15,19]
plt.bar(x,y)
plt.bar(x2,y2,color = "g")
```

```python
# histograms

data = np.array([12,1,15,18,2,49,16,18,19,40,10,74,9,10])
plt.hist(data,bins=10) #bins will present the density or count of numbers bw two points.
plt.title("Histograms Graph")
```

```python
#boxplot -->it will help in percentile

data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True);
```

```python
#pie chart
labels = 'Python' , 'C++' , 'Java'
size = [215,130,200]
color = ['gold', 'yellowgreen' , 'lightcoral']
explode = (0,0,0)

plt.pie(size, explode=explode, labels=labels, colors=color,autopct='%1.1f%%',shadow=False)
plt.axis('equal')
```
